This notebook gives a simple example showing how to use OpenModes. A pair of split-ring resonators are created, and their extinction cross section is calculated

First we load import the `openmodes` packages, along with a few other useful python packages.

```python
# allow plots within the notebook
%pylab inline

# import useful python libraries
import os.path as osp
import numpy as np
import numpy.linalg as la

# import the openmodes packages
import openmodes
from openmodes.visualise import plot_parts
from openmodes.constants import c
```

Then we create a `Simulation` object which holds all the data for the simulation we are going to run

```python
sim = openmodes.Simulation()
```

Next we load the geometry file. This geometry file is written in the scripting language of [gmsh](http://geuz.org/gmsh), a program which converts the geometry into a surface mesh. Here we will use one of the geometry files provided with openmodes. The required mesh density may be specified in the geometry file, but can be over-ridden with the parameter `mesh_tol`

```python
filename = osp.join("geometry", "SRR.geo")
mesh_tol = 2e-3
srr = openmodes.load_parts(filename, mesh_tol)
```

Now we place the parts within the simulation. By default the parts are placed at the origin, so after placement we need to move one of them to the desired distance from the other. It is also possible to rotate parts as well.

```python
srr1 = sim.place_part(srr)
#srr2 = sim.place_part(srr)    

#separation = [20e-3, 0e-3, 0]
#srr2.translate(separation)    
```

To check that the parts have been placed in the correct location, we can visualise them using the provided function. This creates a crude 3D plot, with a cross at each of the nodes of the mesh. It is also possible to produce higher quality 3D plots by exporting to the VTK file format. However, as a quick check, the function `plot_parts` is much simpler, and can be used within a notebook. You can change the perspective by modifying the `view_angles` parameter.

```python
plot_parts(sim.parts, view_angles=(40, 70))
```

We will calculate the extinction cross section for this pair of SRRs, assuming that they are perfectly conducting. We want to excite it with a $y$ polarised plane wave, propagating in the $z$ direction. We will calculate at 201 frequencies between 4 and 8 GHz.

```python
freqs = np.linspace(4e9, 8e9, 201)
e_inc = np.array([0, 1, 0])
k_hat = np.array([0, 0, 1])

# create an empty array to hold extinction data
extinction = np.empty(len(freqs), np.float64)
```

Now loop through all the frequencies. For notational convenience, the time dependence is assumed as $\exp(s t)$ with complex frequency $s$.

At each frequency the impedance matrix is calculated, as is the driving term due to the plane wave. The extinction cross-section $\sigma_{ext}$ can be found from the impedance matrix $Z$ and driving term $V$ as $\sigma_{ext}(s) = V^{*}(s)\cdot Z(s) \cdot V(s)$

```python
for freq_count, freq in enumerate(freqs):
    s = 2j*np.pi*freq
    jk_0 = s/c        
        
    L, S = sim.impedance_matrix(s)
    V = sim.source_term(e_inc, k_hat*jk_0)
    
    extinction[freq_count] = np.dot(V.conj(), la.solve(s*L + S/s, V)).real
```

Now we plot the extinction cross-section as a function of frequency. We normalise the cross section to a disc with the same radius as the SRR to express it in normalised form.

```python
radius = 4e-3

plot(freqs*1e-9, extinction/(pi*radius**2))
xlabel('freq (GHz)')
ylabel('$Q_{ext}$')
title('Extinction effeciency');
```
