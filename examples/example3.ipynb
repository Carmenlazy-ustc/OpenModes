Example 3 - Constructing Scalar Models
--------------------------------------

Once the modes have been found, we can then make use of them to construct scalar models of our meta-atoms.

```python
# allow plots within the notebook, and set some nicer defaults for the plots
from matplotlib.pylab import *
import matplotlib
matplotlib.rcParams['lines.markeredgewidth'] = 2
matplotlib.rcParams['axes.labelsize'] = 14

from IPython.display import Image

# import useful python libraries
import os.path as osp

# import the openmodes package
import openmodes
from openmodes.constants import c, eta_0
```

```python
sim = openmodes.Simulation()
ring = openmodes.load_mesh(osp.join("geometry", "SRR_wide.geo"), mesh_tol=1e-3)
sim.place_part(ring, location=[0e-3, 0, 0]);
```

```python
start_freq = 2e9
start_s = 2j*pi*start_freq

num_modes = 4
mode_s, mode_j = sim.part_singularities(start_s, num_modes);
```

Having found these modes, we can now use them as a basis for constructing models of our resonator which can describe all the dynamics in a relatively simple form. First we generate the scalar model, using the resonant frequencies and the derivative of the eigenimpedance at the resonance.

```python
scalar_models = sim.construct_models(mode_s, mode_j)
```

This model is just a simple scalar impedance, represented by 4 polynomial coefficients as $z_n(s') = a/s' + b + cs' - ds'^2$, in terms of normalised frequency parameter $s' = s/Im(s_n)$. We can inspect the values of these coefficients for each mode

```python
for mode in xrange(num_modes):
    print scalar_models[0][mode].coefficients
```

```python
num_freqs = 101
freqs = np.linspace(1e9, 30e9, num_freqs)
```

Together with the the mode distributions, these form a compact but accurate model of our meta-atom. If we invert these impedance values to obtain admittance, the resonance behaviour is easily observed.

```python
admittance = [1.0/model.scalar_impedance(2j*pi*freqs) for model in scalar_models[0]]

figure(figsize=(14,4))
subplot(1, 4, 1)
plot(freqs*1e-9, admittance[0].real)
plot(freqs*1e-9, admittance[0].imag)
xlim(0, freqs[-1]*1e-9)
legend(['Real', 'Imag'], loc='upper right')
ylabel('Admittance')
xlabel('f (GHz)')
subplot(1, 4, 2)
plot(freqs*1e-9, admittance[1].real)
plot(freqs*1e-9, admittance[1].imag)
xlim(0, freqs[-1]*1e-9)
xlabel('f (GHz)')
subplot(1, 4, 3)
plot(freqs*1e-9, admittance[2].real)
plot(freqs*1e-9, admittance[2].imag)
xlim(0, freqs[-1]*1e-9)
xlabel('f (GHz)')
subplot(1, 4, 4)
plot(freqs*1e-9, admittance[3].real)
plot(freqs*1e-9, admittance[3].imag)
xlim(0, freqs[-1]*1e-9)
xlabel('f (GHz)');
```

Now we utilise this model to calculate the extinction cross section of each mode. First we setup the incident wave polarisation and direction.

```python
e_inc = array([1.0, 1.0, 0], dtype=complex128)/sqrt(2)
k_hat = array([0, 0, 1], dtype=complex128)

p_inc = vdot(e_inc, e_inc).real/eta_0
area = pi*(0.5*ring.max_distance)**2

extinction = empty(num_freqs, np.complex128)
extinction_modes = empty((num_freqs, num_modes), complex128)
```

For comparison purposes we calculate the extinction directly, and time how long this calculation takes. Note that if you increase the mesh density or model a more complex structure, the direct calculation can become very slow, since a large impedance matrix must be filled and solved at every frequency.

```python
%%timeit
for freq_count, freq in enumerate(freqs):
    s = 2j*pi*freq
    Z = sim.calculate_impedance(s)
    V = sim.source_plane_wave(e_inc, k_hat*s/c)
    extinction[freq_count] = np.vdot(V[0], Z[0,0].solve(V[0]))
```

Now we calculate the extinction based on the scalar model for each mode, and also time this calculation

```python
%%timeit
for freq_count, freq in enumerate(freqs):
    s = 2j*pi*freq
    V = sim.source_plane_wave(e_inc, k_hat*s/c)
    extinction_modes[freq_count] = [np.dot(V[0], scalar_models[0][mode].solve(s, V[0])) for mode in xrange(num_modes)]
```

Timings will vary depending on the exact structure to be solved and the performance of the computer. Howeveer in all cases, once the scalar model has been constructed, the system can be solved for different frequencies, input polarisation etc with very low computational cost.

We can now see that the extinction cross-section of each mode can be calculated separately

```python
plot(freqs*1e-9, extinction_modes.real/area/p_inc);
xlim(0, freqs[-1]*1e-9)
xlabel('f (GHz)')
ylabel('$Q_{ext}$')
title('Extinction efficiency of each mode');
```

Note that the fourth mode does not contribute for the chosen incident wave polarisation. The accuracy of this modal decomposition can be checked by comparison with the direct numerical calculation

```python
plot(freqs*1e-9, extinction.real/p_inc/area)
plot(freqs*1e-9, sum(extinction_modes.real, axis=1)/p_inc/area, '+')
legend(['Modal decomposition', 'Direct calculation'])
xlim(0, freqs[-1]*1e-9)
xlabel('f (GHz)')
ylabel('$Q_{ext}$')
title('Extinction efficiency of each mode');
```

The small discrepeancy in this example is because the modes have some small frequency dependence which is neglected in this model. For some structures, this discrepency may be more significant, so checking with the exact calculation is important for validating the approximation.

In addition, we obtain the reactive stored enerrgy of each mode, given by $W_{m}-W_{e}$, the difference between magnetic and electric and stored energies. At resonance these terms go through zero, following from the imaginary part of the admittance.

```python
plot(freqs*1e-9, extinction_modes.imag);
xlim(0, freqs[-1]*1e-9)
xlabel('f (GHz)')
ylabel('$W_m-W_e$')
title('Reactive stored energy');
```
