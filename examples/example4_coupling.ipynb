Example 4 - Calculate Coupling Coefficients
-------------------------------------------
Two split rings are placed in a broadside coupled configuration. The scalar model of each resonator is augmented by additional coefficients representing the coupling between them.

In [D. A. Powell et al, Phys. Rev. B 82, 155128 (2010)](http://dx.doi.org/10.1103/PhysRevB.82.155128) similar coefficients were calculated under the quasi-static approximation for a single mode only. Here the effects of retardation and radiation losses are included, as are the contributions of multiple modes of each ring.

```python
%pylab inline
import matplotlib
matplotlib.rcParams['lines.markeredgewidth'] = 2
matplotlib.rcParams['axes.labelsize'] = 14

import openmodes
from openmodes.constants import c, eta_0
import os.path as osp
```

```python
sim = openmodes.Simulation()
srr = openmodes.load_mesh(osp.join(openmodes.geometry_dir, "SRR_wide.geo"), mesh_tol=1e-3)

srr1 = sim.place_part(srr)
srr2 = sim.place_part(srr, location=[0e-3, 0e-3, 2e-3])
srr2.rotate([0, 0, 1], 180)
```

```python
from openmodes.visualise import plot_parts
plot_parts(sim.parts)
```

```python
start_freq = 2e9
start_s = 2j*pi*start_freq

num_modes = 1
mode_s, mode_j = sim.part_singularities(start_s, num_modes);
```

```python
#freqs = np.linspace(1e9, 10e9, 101)
#freqs = np.linspace(4.8e9, 5.8e9, 101)
freqs = np.linspace(4e9, 12e9, 501)
e_inc = np.array([0, 1, 0], dtype=np.complex128)
k_hat = np.array([1, 0, 0], dtype=np.complex128)

```

```python
extinction_red = empty(len(freqs), np.complex128)
extinction_tot = empty(len(freqs), np.complex128)
extinction_single = empty(len(freqs), np.complex128)

reduced_L = empty((len(freqs), num_modes*2, num_modes*2), np.complex128)
reduced_S = empty((len(freqs), num_modes*2, num_modes*2), np.complex128)
#I_red = np.empty((len(freqs), num_modes*2), np.complex128)

for freq_count, freq in enumerate(freqs):
    s = 2j*np.pi*freq
    
    impedance = sim.impedance(s)
    V = sim.source_plane_wave(e_inc, k_hat*s/c)
    
    #z_mode, mode_j = impedance.eigenmodes(num_modes, use_gram=False)
    #z_mode, mode_j = impedance.eigenmodes(start_j = mode_j)
    impedance_red = impedance.impedance_modes(num_modes, mode_j)
    V_red = impedance.source_modes(V, num_modes, mode_j)

    reduced_L[freq_count] = impedance_red.L
    reduced_S[freq_count] = impedance_red.S
    
    I_red = impedance_red.solve(V_red)
    extinction_red[freq_count] = vdot(V_red, I_red)

    Z_tot, V_tot = impedance.combine_parts(V)
    extinction_tot[freq_count] = vdot(V_tot, Z_tot.solve(V_tot))
    
    extinction_single[freq_count] = vdot(V[0], impedance[0][0].solve(V[0]))
```

```python
# 1 mode
figure(figsize=(12,4))
subplot(121)
plot(freqs*1e-9, extinction_red.real)
plot(freqs*1e-9, extinction_tot.real)
xlim(freqs[0]*1e-9, freqs[-1]*1e-9)
xlabel('f (GHz)')
subplot(122)
plot(freqs*1e-9, extinction_red.imag)
plot(freqs*1e-9, extinction_tot.imag)
xlim(freqs[0]*1e-9, freqs[-1]*1e-9)
xlabel('f (GHz)');
```

```python
# 3 modes
figure(figsize=(12,4))
subplot(121)
plot(freqs*1e-9, extinction_red.real)
plot(freqs*1e-9, extinction_tot.real)
xlim(freqs[0]*1e-9, freqs[-1]*1e-9)
xlabel('f (GHz)')
subplot(122)
plot(freqs*1e-9, extinction_red.imag)
plot(freqs*1e-9, extinction_tot.imag)
xlim(freqs[0]*1e-9, freqs[-1]*1e-9)
xlabel('f (GHz)');
```

```python
area = pi*(srr.max_distance*0.5)**2
incident = vdot(e_inc, e_inc)/eta_0

Q_single = extinction_single/incident/area
Q_pair = extinction_red/incident/area
```

```python
figure(figsize=(12,4))
subplot(121)
plot(freqs*1e-9, Q_pair.real, label='pair')
plot(freqs*1e-9, Q_single.real, label='single')
xlim(freqs[0]*1e-9, freqs[-1]*1e-9)
xlabel('f (GHz)')
ylabel('Extinction efficiency')
legend(loc='upper right')
subplot(122)
plot(freqs*1e-9, Q_pair.imag)
plot(freqs*1e-9, Q_single.imag)
xlim(freqs[0]*1e-9, freqs[-1]*1e-9)
ylabel('Normalised reactance')
xlabel('f (GHz)');
```

```python
plot(freqs*1e-9, reduced_L[:, 0, 0].real)
plot(freqs*1e-9, reduced_L[:, 0, 0].imag)
#ylim (-1e-13, 1e-13)
```

```python
plot(freqs*1e-9, reduced_S[:, 0, 0].real)
plot(freqs*1e-9, reduced_S[:, 0, 0].imag)
```

```python
ss = 2j*pi*freqs
reduced_Z = ss*reduced_L[:, 0, 0] + reduced_S[:, 0, 0]/ss
reduced_Y = 1/reduced_Z
```

```python
plot(freqs*1e-9, reduced_Z.real)
plot(freqs*1e-9, reduced_Z.imag)
```

```python
plot(freqs*1e-9, reduced_Y.real)
plot(freqs*1e-9, reduced_Y.imag)
```

```python
red_L = np.empty_like(reduced_L)
red_S = np.empty_like(reduced_S)

for i in xrange(2*num_modes):
    for j in xrange(2*num_modes):
        red_L[:, i, j] = reduced_L[:, i, j]/sqrt(reduced_L[:, i, i]*reduced_L[:, j, j])
        red_S[:, i, j] = reduced_S[:, i, j]/sqrt(reduced_S[:, i, i]*reduced_S[:, j, j])
```

```python
plot(freqs*1e-9, red_L[:, 0, 3].real)
plot(freqs*1e-9, red_L[:, 0, 3].imag, '--')
#plot(freqs*1e-9, abs(red_L[:, 0, 3:6]))
```

```python
plot(freqs*1e-9, red_S[:, 0, 3].real)
plot(freqs*1e-9, red_S[:, 0, 3].imag, '--')
```
